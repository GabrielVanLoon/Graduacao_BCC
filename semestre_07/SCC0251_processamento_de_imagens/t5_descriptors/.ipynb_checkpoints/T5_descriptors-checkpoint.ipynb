### 1ª Etapa - Leitura dos Parâmetros

O programa recebe como entrada os parâmetros: o nome da imagem com o objeto (`F`), o nome da imagem em que o objeto precisa ser encontrado (`G`) e a quantidade de bits que devem ser considerados para cada pixel (`B`)

```python
# Caso 1: F=wally.png  G=wheres_wally.png  B=3                        OK
# Caso 2: F=blueumbrella.png  G=wheres_blueumbrella.png  B=7          OK
# caso 3: F=jumpingguy.png  G=wheres_jumpingguy.png  B=1              OK
# Caso 4: F=pokeball.png  G=wheres_pokeball.png  B=1                  OK
# Caso 5: F=pumpkin.png  G=wheres_pumpkin.png  B=8
F = input().rstrip()
G = input().rstrip()
B = int(input().rstrip())
```

### 2ª Etapa - Pré-processamento

Antes de iniciar a criação dos descriptors vamos primeiro transformar as imagens do RGB para grayscale utilizando o método de Luminance e também remover os (8-B) bits menos significativos.

```python
def luminance_grayscale(image):
    float_image = image.astype(np.float32)
    return np.floor(0.299*float_image[:,:,0] + 0.587*float_image[:,:,1] + 0.114*float_image[:,:,2]).astype(np.uint8)

def image_bitshift(image, bits=8):
    int_image = image.astype(np.uint8)
    return (int_image >> (8-bits))
```

```python
f_gray_image = image_bitshift(luminance_grayscale(f_image), B)
g_gray_image = image_bitshift(luminance_grayscale(g_image), B)
```

### 2ª Etapa - Aplicando os Descriptors
 
- **Normalised Histogram**
- **Haralick (Co-ocurrence parameters)**
- **Histogram of Oriented Gradients**

```python
def desc_normalised_histogram(image, bits=8):
    """
    Computa e retorna um histograma de 2**bits posicoes normalizado. É assumido que a imagem
    enviada está no formato np.uint8 e já shiftada para a quantidade de bits correta.
    """
    hist, _ = np.histogram(image, bins=2**bits, range=(0,2**bits))
    hist    = hist.astype(np.float32) / float(image.size)
    return hist/np.linalg.norm(hist) if np.linalg.norm(hist) != 0.0 else hist
```

```python
def desc_haralick(image, bits=8):
    """
    Computa e retorna o vetor descritor utilizando o método de Haralick. É assumido que aimagem
    está no formato np.uint8 e já shiftada para a quantidade de bits correta.
    """
    N, M = image.shape
    
    # First step - calculating the co-ocurrence matrix
    C = np.zeros((2**bits,2**bits), dtype=np.float32)
    for i in range(image.shape[0]-1):
        for j in range(image.shape[1]-1):
            C[image[i,j],image[i+1,j+1]] += 1.0   
    
    C = C/np.sum(C)
    
    # Helpers variables used in second step
    i_index, j_index = np.indices((2**bits,2**bits)) 
    diff_index = i_index - j_index 
    
    mi_i = np.sum(i_index * C)
    mi_j = np.sum(j_index * C)
    
    di = np.sum(np.power(i_index - mi_i, 2) * C)
    dj = np.sum(np.power(j_index - mi_j, 2) * C)
    
    # Second step - calc. the haralick parameters (energy, entropy, contrast, correlation and homogeneity)
    energy   =  np.sum(np.power(C, 2))
    entropy  = -np.sum(C * np.log(C + 0.001))
    contrast =  np.sum(np.power(diff_index,2) * C) / C.size
    correlation = (np.sum((i_index * j_index * C)) - mi_i*mi_j)/(di * dj) if (di>0.0 and dj>0.0) else 0.0 
    homogeneity = np.sum(C / (1 + np.abs(i_index - j_index)))
    
    dt = np.array([energy, entropy, contrast, correlation, homogeneity], dtype=np.float32)
    return dt/np.linalg.norm(dt) if np.linalg.norm(dt) != 0.0 else dt
```

```python
def desc_histogram_oriented_grads(image):
    """
    Computa e retorna o vetor descritor utilizando o método de histograma de gradientes. 
    É assumido que aimagem está no formato np.uint8 e já shiftada para a quantidade de bits correta.
    """
    np.seterr(divide='ignore', invalid='ignore')
    
    # First step - Get gradient matrixes
    ws_x = np.array([[-1,-2,-1],[0,0,0],[1,2,1]], dtype=int)
    ws_y = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], dtype=int)
    
    gx = convolve(image, ws_x).astype(np.float32)
    gy = convolve(image, ws_y).astype(np.float32)
    
    # Second step - get magnitude (M) and angle (Fi)
    M =  np.sqrt(np.power(gx,2) + np.power(gy,2))
    M /= np.sum(M) if np.sum(M) != 0.0 else 1.0
    
    # Third step - get the normalized descriptor
    arctan_m = np.arctan(np.nan_to_num(gy/gx, nan=np.Infinity)) #Corner case 0/0 -> nan set to 
    Fi = np.degrees(arctan_m + (np.pi/2.0)) 
    Fi = np.where(Fi==180.0, 179.0, Fi) # Workaround change 180 to 179 to use np.histogram :D
    Fi_hist, _ = np.histogram(Fi, bins=9, range=(0,180), weights=M)
    
    return Fi_hist/np.linalg.norm(Fi_hist) if np.linalg.norm(Fi_hist) != 0.0 else Fi_hist
```

```python
# Generating descriptor of f
f_norm_hist = desc_normalised_histogram(f_gray_image, B)
f_haralick = desc_haralick(f_gray_image, B)
f_grads_hist = desc_histogram_oriented_grads(f_gray_image)

f_descriptor = np.concatenate((f_norm_hist, f_haralick, f_grads_hist), axis=None)
```

```python
# Divide image G in windows of 32x32 (stride 16) and for each window:
# - Calculate the subimage descriptor
# - Calculate the distance to f_descriptor

# Helpers index matrix to iterate the windows with stride 32
i_index, j_index = np.indices(g_gray_image.shape) 
i_index = i_index[0:(g_gray_image.shape[0]-31):16, 0:(g_gray_image.shape[1]-31):16]
j_index = j_index[0:(g_gray_image.shape[0]-31):16, 0:(g_gray_image.shape[1]-31):16]

# Distance matrix and other importante values
D = np.zeros(i_index.shape, dtype=np.float32)
W = 32 
```

```python
# Running all the windows
for i in range(D.shape[0]):
    for j in range(D.shape[1]):
        window = g_gray_image[i_index[i,j]:i_index[i,j]+W, j_index[i,j]:j_index[i,j]+W]
            
        w_norm_hist  = desc_normalised_histogram(window, B)
        w_haralick   = desc_haralick(window, B)
        w_grads_hist = desc_histogram_oriented_grads(window)
        w_descriptor = np.concatenate((w_norm_hist, w_haralick, w_grads_hist), axis=None)
        
        # Distance from window descriptor to f
        D[i,j] = np.sqrt(np.sum(np.power(w_descriptor-f_descriptor,2)))
        
        # Checking for errors in descriptors
        #if np.isnan(np.min(w_descriptor)):
        #   print(i,j)
        #   if np.isnan(np.min(w_norm_hist)):
        #       print('Norm', w_norm_hist)
        #   if np.isnan(np.min(w_haralick)):
        #       print('Haralick', w_haralick)
        #   if np.isnan(np.min(w_grads_hist)):
        #       print('Grads', w_grads_hist)
                
        if (i,j) == (1,11):
            print('1-11', D[i,j])
            print('Norm D:', np.sqrt(np.sum(np.power(w_norm_hist-f_norm_hist,2))))
            print('Haralick D:', np.sqrt(np.sum(np.power(w_haralick-f_haralick,2))))
            print('Grads D:', np.sqrt(np.sum(np.power(w_grads_hist-f_grads_hist,2))))
        elif (i,j) == (6,11):
            print('6-11', D[i,j])
            print('Norm D:', np.sqrt(np.sum(np.power(w_norm_hist-f_norm_hist,2))))
            print('Haralick D:', np.sqrt(np.sum(np.power(w_haralick-f_haralick,2))))
            print('Grads D:', np.sqrt(np.sum(np.power(w_grads_hist-f_grads_hist,2))))
```
