### 2ª Etapa - Preparando Funções de Base

A seguir, iremos implementar as funções de realce que, na prática são duas: 

- **Nenhum Realce** (opção 0)
- **Realce Baseado em Histograma**
    - (opção 1) para cada uma das `imglow` calcula o histograma e aplica a função de realce
    - (opção 2) calcula o histograma conjunto das 4 `imglow` e depois aplica o mesmo realce em todas
- **Correção Gamma**
    - (opção 3) realiza a técnica de realce em todas as imagens utilizando o parâmetro `G`

Vamos, portanto, definir primeiro alguma das funções que serão necessárias para o cálculo dos histogramas e realces, iniciando pela função que irá computar o vetor de distribuição dos pixels de uma dada imagem (histograma).

```python
def get_pixel_histogram(image, normalize=True, cumulative=True, max_range=256):
    histogram = np.zeros(max_range)
    
    for i in range(max_range):
        histogram[i] = np.sum(image == i)
        
        if cumulative and i > 0:
            histogram[i] += histogram[i-1]

    if normalize:
        histogram = histogram / imghigh.size
        
    return histogram
```

Em seguida vamos criar a função responsável por realizar o realce utilizando o histograma. Esse método é também chamado de  **Histogram Equalisation**. Nesse método nos utilizamos a distribuição original para gerar uma nova distribuição e, com isso, mapear os valores da imagem original para novos valores. O realce ocorre por meio da seguinte equação: 

$$ s = T(z) = \frac{L-1}{M N} hc(z)$$

```python
def histogram_equalisation(image, histogram, max_range=256):
    eq_image = np.zeros(image.shape)
    
    for z in range(max_range):
        s = ((max_range-1)/image.size) * histogram[z]
        eq_image[np.where(image == z)] = s
    
    eq_image = eq_image.astype(np.uint8)
    return eq_image    
```

Por fim, o último metodo de realce chamado de **Correção Gamma** percorre a imagem e atualiza cada pixel seguindo a seguinte fórmula de correção:

$$ L_i(x,y) = floor(255 * (\frac{L_i(x,y)}{255})^\frac{1}{\gamma}           )$$

```python
def gamma_correction(image, gamma=0.5):
    new_image = image.copy().astype(float)

    new_image = 255.0 * np.power(new_image/255.0, 1.0/gamma)

    new_image = new_image.astype(np.uint8)
    return new_image
```

### 3ª Etapa - Preparando as Opções de Realce

A seguir definimos as 4 opções de funções de realce que podem ser aplicadas nas imagens de baixa resolução

```python
def opt0_no_enhance():
    # Apply nothing, keep the images as init
    return False

def opt1_single_histogram_enhance():
    # Apply histogram equalisation for each image
    for i in range(4):
        hist      = get_pixel_histogram(imglow[i], normalize=False, cumulative=True)
        imglow[i] = histogram_equalisation(imglow[i], hist)
        
def opt2_set_histogram_enhance():
    # Calculate a unique histogram from the set of images
    hist = 0.0
    for i in range(4):
        hist = hist + get_pixel_histogram(imglow[i], normalize=False, cumulative=True)
    
    # Divide the histogram to get the average in each distribution
    hist = (hist/4.0).astype(int)
    
    # Apply the equalisation for all images
    for i in range(4):
        imglow[i] = histogram_equalisation(imglow[i], hist)
        
def opt3_gamma_corrections():
    for i in range(4):
        imglow[i] = gamma_correction(imglow[i], gamma=G)
        
f_options = [ opt0_no_enhance, opt1_single_histogram_enhance, opt2_set_histogram_enhance, opt3_gamma_corrections ]    
```

Por fim, aplicamos o realce selecionado nas imagens

```python
f_options[F]()
```

### 4ª Etapa - Superresolução

Por fim, vamos utilizar as 4 imagens de baixa resolução para compor uma única imagem de tamanho $(2N,2N)$ utilizando um método simples que mapeia cada pixel de uma imagem para uma região equivalente na imagem final da seguinte forma:

$$ Região(x,y) = \begin{matrix}
px_1 & px_2 \\
px_3 & px_4
\end{matrix}$$

```python
imgsuper = np.zeros(imghigh.shape)
# print(imgsuper)

imgsuper[0::2,0::2] = imglow[0]
imgsuper[0::2,1::2] = imglow[1]
imgsuper[1::2,0::2] = imglow[2]
imgsuper[1::2,1::2] = imglow[3]
```

### 5ª Etapa - Comparando com a Referência

No primeiro parâmetro lido recebemos o caminho da imagem de referência que deveria ser resultante do caso de teste. Nesta etapa vamos calcular a Média da Raiz do Erro Quadrático (RMSE) para verificar quão próximo chegamos do resultado correto.

```python
# print(imgsuper)
# print(imghigh)
```

```python
imgsuper = imgsuper.astype(int)
imghigh  = imghigh.astype(int)

# Calculating the error 
error =  np.sqrt(np.sum(np.power(imghigh-imgsuper, 2.0))/imghigh.size)
print(format(error, ".4f"))
```
